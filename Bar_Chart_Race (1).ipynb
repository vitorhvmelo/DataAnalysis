```
import pandas as pd
import bar_chart_race as bcr
```

```
df = pd.read_excel('/content/drive/My Drive/Data Analysis /HIST_PAINEL_COVIDBR_08ago2020.xlsx')
```

```
df.head() 
```

```
df.dropna(subset=["municipio"], inplace=True) #Dropamos inplace o na pela coluna municipio
```

```
df.head()
```

#Por regiao do Brasil
criamos um df com o groupby da regiao e data


```
dfAcumulado = pd.DataFrame(df.groupby(["regiao","data"]).sum()["obitosAcumulado"]).unstack().T.droplevel(level=0)
```

```
dfAcumulado
```

```
def summary(values, ranks):
    total_deaths = values.sum()
    s = f'Total de Mortes - {total_deaths:,.0f}'
    return {'x': .99, 'y': .05, 's': s, 'ha': 'right', 'size': 8}
bcr.bar_chart_race(dfAcumulado, period_summary_func=summary)
```
